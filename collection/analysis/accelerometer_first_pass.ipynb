```python
import pandas as pd
import numpy as np
import glob
```

```python
acc = glob.glob("accurate*")

def combine_csv(files, output_file):
    header_saved = False
    with open(output_file,'wb') as fout:
        for file_want in files:
            with open(file_want) as fin:
                header = next(fin)
                if not header_saved:
                    fout.write(header)
                    header_saved = True
                for line in fin:
                    fout.write(line)
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
mover = moving_average(belly_one_comb)
```
